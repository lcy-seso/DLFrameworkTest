```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

palette = 'Set2'
```

```python
df = pd.read_table(
    '30000_1024_merged.tsv',
    header=0,
    sep='\t',
    usecols=['Method', 'Launch Config', 'Elapsed Time(ms)','Bandwidth Utilization(%)'])

g = sns.catplot(
    data=df,
    kind='bar',
    x='Launch Config',
    y='Bandwidth Utilization(%)',
    hue='Method',
    ci='sd',
    palette=palette,
    alpha=.6,
    height=6).set(title='matrix shape = [30000, 1024]')
g.set_axis_labels('Launch Config',
                  'Bandwidth Utilization')
g.set_xticklabels(size=12)
g.set_yticklabels(size=12)

g._legend.remove()

plt.tight_layout()
plt.xticks(rotation=60)
plt.legend(
    bbox_to_anchor=(0.01, 0.99),
    loc='center right',
    fontsize='12',
    title_fontsize='16')

fig = g.fig
fig.show()

fig.savefig('access_test1.png', dpi=500, bbox_inches='tight')
```
